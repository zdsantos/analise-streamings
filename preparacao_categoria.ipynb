```python
import pandas as pd
df = pd.read_csv("./data/disney_plus_titles.csv")

df.head()
```

Quantidade de linhas nulas em cada coluna

```python
df.shape
```

```python
def split_column_values(row, column_name):
    return list(map(lambda c: c.strip(), row[column_name].split(sep=",")))
```

```python
df_separate_cat = pd.DataFrame(columns= df.columns);

for index, row in df.iterrows():
    values_list = split_column_values(row, "listed_in")
    for cat in values_list:
        new_row = row
        new_row["listed_in"] = cat
        df_separate_cat = pd.concat([df_separate_cat, new_row.to_frame().T], ignore_index=True)
```

```python
df_separate_cat.to_csv("./data/disney_plus_titles_separeted.csv")
```
