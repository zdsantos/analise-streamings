```python
import pandas as pd
file_name = "./data/base_unificada"
df = pd.read_csv(file_name + ".csv");
```

```python
import numpy as np

def split_column_values(row, column_name):
    value = row[column_name]
    if value and value != "" and value is not np.nan:
        return list(map(lambda c: c.strip(), value.split(sep=",")))
```

```python
df_separate_cat = pd.DataFrame(columns= df.columns);

column_name = "country"

for index, row in df.iterrows():
    values_list = split_column_values(row, column_name)
    if values_list != None:
        for value in values_list:
            new_row = row
            new_row[column_name] = value
            df_separate_cat = pd.concat([df_separate_cat, new_row.to_frame().T], ignore_index=True)
    else:
        df_separate_cat = pd.concat([df_separate_cat, row.to_frame().T], ignore_index=True)
```
